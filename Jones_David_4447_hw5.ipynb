```python
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
import numpy as np 
import csv
import urllib
import time

#response = requests.get('https://www.indeed.com/jobs?q=data+scientist&l=CO&start=0')

url = "https://www.indeed.com/jobs?q=data+scientist&l=CO&start="
jobTitle = []
companyName = []
companyLocation = []
summary = []
jobID = []

for i in range(0,100,10):
    urls = str(url) + str(i)
    page = requests.get(urls)
    soups = soup(page.text,'html.parser')
    card = soups.find_all('div', 'slider_container')
            
    for i in range(len(card)):
        for name in card[i].find_all('span',title=True):
            jobTitle.append(name['title'])
            cn = card[i].find('span',"companyName").text
            companyName.append(cn)
            cl = card[i].find('div',"companyLocation").text
            companyLocation.append(cl)
            s = card[i].find('div',"job-snippet").text
            summary.append(s)
    
#     for job in soups.find_all('div',{"id":"mosaic-provider-jobcards"}):  
#         for post in job.find_all('a', 'data-jk'==True):
            
#             if post.find('a', 'data-jk'==True):
#                 job_desc_href = urls+"&vjk="+str(post)
#                 idpage = requests.get(job_desc_href)
#                 idSoup = soup(idpage.text, 'html.parser')
#                 job_desc = idSoup.find(id='jobDescriptionText')
#                 jobID.append(job_desc)

# I wasn't able to get the individual description request to work, but I feel like I was close to figuring it out.
# Moving forward, I'm going to use the provied .pkl file for the remainder of the assignment.
        
DATA = {'Job':jobTitle , 'Company':companyName , 'Location':companyLocation , 'Summary':summary}
df = pd.DataFrame(DATA) 

df
```

# Q 3 b(1.5 point) - Top 3 most demanding skills(like Python, AWS, SQL ...)



```python
values = []

columns = ['Python', 'SQL', 'AWS', 'RESTFUL', 'Machine_Learning', 'Deep_Learning', 'Text_Mining', 
        'NLP', 'SAS', 'Tableau', 'Sagemaker', 'TensorFlow', 'Spark']

for i in columns:
    try: 
        values.append(job_df[i].value_counts()[1])
    except KeyError:
        values.append(0) 

columnData= {'skill': columns, 'value': values}
columnDF = pd.DataFrame(columnData)

columnDF.sort_values(by=['value'], ascending = False)
columnDF.head(3)
```
